```python
Country_name = X.Country.value_counts().index
```

```python
ratings = X.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"rating count"})
ratings
```

Observations-
1.When Rating is between 4.5 to 4.9 = Excellent

2.When Rating is between 4.0 to 4.4 = Very Good

3.When Rating is between 3.5 to 3.9 = Good

4.When Rating is between 2.5 to 3.4 = Average

5.When Rating is between 1.8 to 2.4 = Poor

```python
import seaborn as sb
sb.barplot(x=ratings['Aggregate rating'],y=ratings['rating count'])
```

Observations-
1.not rated count is very high     
2.Maximum number rating are between 2.5 to 3.4
